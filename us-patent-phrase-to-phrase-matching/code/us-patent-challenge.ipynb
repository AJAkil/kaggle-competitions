```python
import pickle
import pandas as pd
import os
import numpy as np
```

```python
with open('../input/cpc-titles-pickle/cpc_titles.pickle','rb') as handle:
    cpc_titles = pickle.load(handle)
```

```python
train_df = pd.read_csv('../input/us-patent-phrase-to-phrase-matching/train.csv')
test_df = pd.read_csv('../input/us-patent-phrase-to-phrase-matching/test.csv')
```

```python
test_df.head()
```

```python
def get_cpc_details(context_codes):
    return cpc_titles[context_codes].lower()
```

```python
train_df['context_details'] = train_df['context'].apply(lambda x: get_cpc_details(x))
test_df['context_details'] = test_df['context'].apply(lambda x: get_cpc_details(x))
```

```python
test_df.iloc[0:1]['context_details']
```

```python
train_df['input'] = train_df['anchor'] + '[SEP]' + train_df['context_details']
```
